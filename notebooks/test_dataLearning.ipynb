```python
import sys
sys.path.append('../api')

from datalearning import DataLearning
from rdflib import URIRef, BNode
```

```python
datafile = "/home/dmlung/workspace/ddi-cdi-xas/cdi-xas/resources/prov_generated_by.jsonld"
configpath = "/home/dmlung/workspace/ddi-cdi-xas/cdi-xas/resources/"
data = DataLearning(configpath, datafile, format="json-ld")
data.load_data()
data.load_cdi()
```

```python
def learn(field: str):
    triples: tuple[str,str,str] = data.configurate_triples(field, data.g)
    for s,_,_ in triples:
        if (isinstance(s, URIRef) and s not in data.nodes):
            inner_triples: tuple[str,str,str] = data.get_related_triples(data.g, s)
            for _,_,i_o in inner_triples:
                learn(i_o)
```

```python
field = "http://schema.org/Dataset"
triples: tuple[str,str,str] = data.configurate_triples(field, data.g)
for s,_,_ in triples:
    learn(s)

with open(configpath + 'out/' + 'test_schema.nt', 'w') as out_file:
    out_file.write(data.schemagraph.serialize(format="nt")) 

with open(configpath + 'out/' + 'test_schema.jsonld', 'w') as out_file:
    out_file.write(data.schemagraph.serialize(format="json-ld", context=data.schemagraph.context))
```
